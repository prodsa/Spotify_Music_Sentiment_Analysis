```python
#importing seaborn package for visualization
import seaborn as sns

#creating a df with audio features (numeric attributes)
audio = spotify[['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'key', 'liveness', 'loudness', 'mode',
       'speechiness', 'tempo', 'time_signature', 'valence']]
sns.kdeplot(audio);
```

```python
#creating jointplot with energy and valence
with sns.axes_style('white'):
    sns.jointplot("energy", "valence", audio, kind='hex')
```
